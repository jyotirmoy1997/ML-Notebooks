### Importing Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Social_Network_Ads.csv")

df
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

X, y
```

### Splitting the data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
X_train, X_test, y_train, y_test
```

### Performing Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

X_train, X_test
```

### Training the model

```python
from sklearn.ensemble import RandomForestClassifier

classifier1 = RandomForestClassifier(n_estimators=70, criterion='entropy')

classifier1.fit(X_train, y_train)
```

### Predicting the test result

```python
y_pred1 = classifier1.predict(X_test)
y_pred1
```

```python
from sklearn.linear_model import LogisticRegression

classifier2 = LogisticRegression()
classifier2.fit(X_train, y_train)
```

```python
y_pred2 = classifier2.predict(X_test)
y_pred2
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, RocCurveDisplay

auc1 = roc_auc_score(y_test, y_pred1)
auc2 = roc_auc_score(y_test, y_pred2)

auc1, auc2
```

```python
fpr1, tpr1, thresholds1 = roc_curve(y_test, y_pred1)
fpr2, tpr2, thresholds2 = roc_curve(y_test, y_pred2)

plt.figure()
plt.plot(fpr1, tpr1, label=f'Random Forest')
plt.plot(fpr2, tpr2, label=f'Logistic Regression')

plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(True)
plt.show()
```
