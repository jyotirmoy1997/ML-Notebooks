### Importing Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Mall_Customers.csv")

df
```

```python
X = df.iloc[:, [3, 4]].values
X
```

### Plotting the Dendrogram

```python
import scipy.cluster.hierarchy as hc

dendrogram = hc.dendrogram(hc.linkage(X, method='ward'))
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidena Distance")
plt.show()
```

### Training the model with optimum number of clusters found from Dendrogram

```python
from sklearn.cluster import AgglomerativeClustering

hierchical_cluster = AgglomerativeClustering(n_clusters=5, linkage='ward')
y_hc = hierchical_cluster.fit_predict(X)

y_hc
```
