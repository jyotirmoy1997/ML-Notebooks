### Importing and Reading the Dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Position_Salaries.csv")

df
```

```python
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values

X, y
```

### Transforming Data to Polynomial

```python
from sklearn.preprocessing import PolynomialFeatures

poly_fet = PolynomialFeatures(degree=5)
X_poly = poly_fet.fit_transform(X)
X_poly
```

### Training the model using polynomial data

```python
from sklearn.linear_model import LinearRegression

poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
```

### Predicting new test data
* First converting the new data to match dimensions

```python
x_new = np.array([[11]])
x_new = poly_fet.fit_transform(x_new)
```

* Making the predictions

```python
poly_reg.predict(x_new)
```

### Plotting the graph to see the eficacy of the model

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X, y, color="red")
plt.plot(X, poly_reg.predict(poly_fet.fit_transform(X)), color="blue")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```
